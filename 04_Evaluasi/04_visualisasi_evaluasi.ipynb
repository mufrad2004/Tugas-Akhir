```python
import matplotlib.pyplot as plt

# =========================
# Data (Evaluasi Terbaru)
# =========================
topologies = [50, 100]

# Rule-Based (selalu sempurna)
rb_subset       = [1.0, 1.0]
rb_hamming      = [1.0, 1.0]
rb_micro_recall = [1.0, 1.0]
rb_macro_f1     = [1.0, 1.0]

# Random Forest (Cross-Validation)  hasil terbaru
rf_subset       = [0.91, 0.9287]
rf_hamming      = [0.9908, 0.9929]
rf_micro_recall = [0.5843, 0.7083]
rf_macro_f1     = [0.6893, 0.8484]

# =========================
# Plot
# =========================
fig, axes = plt.subplots(1, 4, figsize=(16, 4), sharex=True)
fig.suptitle("Perbandingan Kinerja Rule-Based dan Random Forest\n(50 vs 100 Topologi)  Dataset Terbaru", fontsize=12)

# 1. Subset Accuracy
ax = axes[0]
ax.plot(topologies, rb_subset, marker="o", label="Rule-Based")
ax.plot(topologies, rf_subset, marker="o", label="Random Forest")
ax.set_title("Subset Accuracy")
ax.set_xlabel("Jumlah Topologi")
ax.set_ylabel("Nilai")
ax.set_ylim(0, 1.05)
ax.grid(True)

# 2. Hamming Accuracy
ax = axes[1]
ax.plot(topologies, rb_hamming, marker="o", label="Rule-Based")
ax.plot(topologies, rf_hamming, marker="o", label="Random Forest")
ax.set_title("Hamming Accuracy")
ax.set_xlabel("Jumlah Topologi")
ax.set_ylim(0, 1.05)
ax.grid(True)

# 3. Micro Recall
ax = axes[2]
ax.plot(topologies, rb_micro_recall, marker="o", label="Rule-Based")
ax.plot(topologies, rf_micro_recall, marker="o", label="Random Forest")
ax.set_title("Micro Recall")
ax.set_xlabel("Jumlah Topologi")
ax.set_ylim(0, 1.05)
ax.grid(True)

# 4. Macro F1-Score
ax = axes[3]
ax.plot(topologies, rb_macro_f1, marker="o", label="Rule-Based")
ax.plot(topologies, rf_macro_f1, marker="o", label="Random Forest")
ax.set_title("Macro F1-Score")
ax.set_xlabel("Jumlah Topologi")
ax.set_ylim(0, 1.05)
ax.grid(True)

# Legend (satu saja di luar subplot)
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=2, bbox_to_anchor=(0.5, 0.02))

plt.tight_layout(rect=[0, 0.05, 1, 0.9])
plt.show()
```
